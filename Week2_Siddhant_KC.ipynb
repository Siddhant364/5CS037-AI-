3. For the column with missing values fill the values using various techniques we discussed above.
Try to explain why did you select the particular methods for particular column.

```python
import pandas as pd
df = pd.read_csv("drive/MyDrive/Dataset/medical_students_dataset.csv")
df["Student ID"] = df["Student ID"].ffill()
df["Student ID"] = df["Student ID"].bfill()

df["Age"] = df["Age"].fillna(df["Age"].median())

df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])

df["Height"] = df["Height"].fillna(df["Height"].mean())

df["Weight"] = df["Weight"].fillna(df["Weight"].mean())

df["Blood Type"] = df["Blood Type"].fillna(df["Blood Type"].mode()[0])

df["BMI"] = df["BMI"].fillna(df["BMI"].median())

df["Temperature"] = df["Temperature"].fillna(df["Temperature"].mean())

df["Heart Rate"] = df["Heart Rate"].fillna(df["Heart Rate"].median())

df["Blood Pressure"] = df["Blood Pressure"].interpolate()

df["Cholesterol"] = df["Cholesterol"].fillna(df["Cholesterol"].median())

df["Diabetes"] = df["Diabetes"].fillna(df["Diabetes"].mode()[0])

df["Smoking"] = df["Smoking"].fillna(df["Smoking"].mode()[0])

df.isnull().sum()
```

**Problem 3**

```python
import pandas as pd
df = pd.read_csv("drive/MyDrive/Dataset/Titanic-Dataset.csv")

# 1. Create one-hot encoded columns for 'Embarked'
embarked_dummies = pd.get_dummies(df["Embarked"], prefix="Embarked")

# 2. Add them to the original DataFrame
df = pd.concat([df, embarked_dummies], axis=1)

# 3. Drop the original 'Embarked' column
df = df.drop("Embarked", axis=1)

# 4. View the modified DataFrame
print(df.head())
```

**Problem 4**

```python
import matplotlib.pyplot as plt
mean_survival = df.groupby("Sex")["Survived"].mean()
print(mean_survival)
plt.figure(figsize=(7,5))
df.boxplot(column="Survived", by="Sex")

plt.title("Survival Distribution by Gender")
plt.suptitle("")      # removes automatic extra title
plt.xlabel("Sex")
plt.ylabel("Survived (0 = No, 1 = Yes)")
plt.show()
```

**Problem 5**

```python
df = pd.get_dummies(df, columns=["Embarked"], prefix="Embarked")
df["Embarked"] = (
    df["Embarked_C"].replace({1: "C"}) +
    df["Embarked_Q"].replace({1: "Q"}) +
    df["Embarked_S"].replace({1: "S"})
)


plt.figure(figsize=(8,5))

df.boxplot(column="Survived", by="Embarked")

plt.title("Survival Distribution by Port of Embarkation")
plt.suptitle("")
plt.xlabel("Port of Embarkation")
plt.ylabel("Survived (0 = No, 1 = Yes)")

plt.show()

```
