# **To - Do - 6:**

Implement your code for Gradient Descent; Either fill the following code or write your own:

```python
import numpy as np

def gradient_descent(X, Y, W, alpha, iterations):
    m = X.shape[0]
    cost_history = []

    for _ in range(iterations):

        Y_pred = np.dot(X, W)

        error = Y_pred - Y

        dw = (1 / m) * np.dot(X.T, error)

        W = W - alpha * dw

        cost = (1 / (2 * m)) * np.sum(error ** 2)
        cost_history.append(cost)

    return W, cost_history

```

# **To - Do - 8:**

Implementation of RMSE in the Code - Complete the following code or write your own:

```python
import numpy as np
def rmse(Y, Y_pred):
    n = len(Y)
    error = Y - Y_pred
    rmse = np.sqrt((1 / n) * np.sum(error ** 2))

    return rmse
```

# **To - Do - 9 - Implementation in the Code:**

Complete the following code or write your own for r2 loss:

```python
import numpy as np

def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    r2 = 1 - (ss_res / ss_tot)

    return r2
```

```python
def main():
    # Load dataset
    data = pd.read_csv('drive/MyDrive/Dataset/student.csv')

    X = data[['Math', 'Reading']].values
    Y = data['Writing'].values

    # Shuffle and split
    np.random.seed(42)
    indices = np.arange(len(X))
    np.random.shuffle(indices)
    split = int(0.8 * len(X))

    X_train = X[indices[:split]]
    X_test  = X[indices[split:]]
    Y_train = Y[indices[:split]]
    Y_test  = Y[indices[split:]]

    # Initial weights and iterations
    W = np.zeros(X_train.shape[1])
    iterations = 1000

    # Experiment with different learning rates
    alphas = [0.000001, 0.00001, 0.0001]

    for alpha in alphas:
        W_opt, cost_history = gradient_descent(X_train, Y_train, W, alpha, iterations)

        Y_train_pred = np.dot(X_train, W_opt)
        Y_test_pred  = np.dot(X_test, W_opt)

        print(f"\nLearning Rate: {alpha}")
        print("Final Weights:", W_opt)
        print("Train RMSE:", rmse(Y_train, Y_train_pred))
        print("Test RMSE:", rmse(Y_test, Y_test_pred))
        print("Train RÂ²:", r2(Y_train, Y_train_pred))
        print("Test RÂ²:", r2(Y_test, Y_test_pred))


if __name__ == "__main__":
    main()
```

Based on the gradient descent experiments with different learning rates, the model shows varying performance. Increasing the learning rate to 0.00001 improves performance, reducing Train and Test RMSE to 5.16 and 5.26, and increasing RÂ² scores to 0.89 (train) and 0.87 (test), indicating better learning and generalization. A further increase to 0.0001 results in the best performance among the tested rates, with Train RMSE of 4.58, Test RMSE of 4.61, and RÂ² scores of 0.91 (train) and 0.90 (test). Overall, the model does not overfit, as train and test errors are close, and the performance is acceptable. The optimal learning rate for this dataset appears to be 0.0001, which balances fast convergence and stable, accurate predictions.
